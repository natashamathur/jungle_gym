```python
import pandas as pd
import jellyfish as jf
```

```python
main = pd.read_table('SF-Rest-SmallSample.tsv', header=0)
small = main.head(10000)
```

```python
def isNaN(x):
    return x != x
```

```python
def friday(df):
    matches = {}
    pcs = (df.business_postal_code.unique())
    for pc in pcs:
        pc_df = df[df['business_postal_code'] == pc]
        print("still alive", pc, pc_df.shape)
        x = pc_df.drop_duplicates(subset=['business_name', 'business_address'])
        matches = fri_loop(x, matches)
    return matches
```

```python
def fri_loop(df, matches):
    for idx_r, r in df.iterrows():
        for idk_j, j in df.iterrows():
            #print(r.busines_name, j.business_name)
            s = get_score_df(r,j)
            if s:
                if r.business_name not in matches.keys():
                    matches[j.business_name] = [r.business_name, r.business_address, r.business_city, r.business_state,
               r.business_postal_code, r.business_latitude, r.business_longitude,
               r.business_phone_number]
    return matches
```

```python
def get_score_df(i, j):
        if i[0] == j[0]:
            return False
            
        i = list(i)
        i = i[1:]
        i = [str(a) for a in i]
        j = list(j)
        j = j[1:]
        j = [str(a) for a in j]

        b = jf.jaro_winkler(i[0], j[0])
        a = jf.jaro_winkler(i[1], j[1])
        pc = jf.jaro_winkler(str(i[4]), str(j[4]))
        
        if isNaN(i[7]) or isNaN(j[7]):
            #print("there is a nan")
            score = b*0.5 + a*0.45 + pc*0.05
        else:
            try:
                ph_n = jf.jaro_winkler(i[7], j[7])
                score = b*0.4 + a*0.4 + ph_n*0.15 + pc*0.05
            except:
                score = b*0.5 + a*0.45 + pc*0.05
            
        if score > 0.9:
            return True
        return False
```

```python
m = friday(main)
```

```python
def trial(row, matches = m, test_df = test_df):
    if row.business_name in matches.keys():
        tr = [row.business_name, row.business_address, row.business_city, row.business_state,
               row.business_postal_code, row.business_latitude, row.business_longitude,
               row.business_phone_number]
        v = list(matches[row.business_name])
        test_df = test_df.replace(tr,v)
    return row
```
